```python
df['cp'].value_counts()
```

```python
df['cp']=df['cp'].astype('str')
```

```python
def category(x):
    for i in x:
        if i=='0':
            return 'Cat0'
        elif i=='1':
            return 'Cat1'
        elif i=='2':
            return 'Cat2'
        else:
            return 'Cat3'
df['cpcategory']=df['cp'].apply(lambda x : category(x))
```

```python
df.head()
```

```python
ax = sns.countplot(df['cp'],palette='viridis')
plt.yticks([0,1,2,3],labels=('Cat0','Cat1','Cat2','Cat3'))
plt.grid()
for item in ax.containers:
    ax.bar_label(item)
```

```python
sns.countplot(x='cpcategory',hue='target',data=df)
plt.legend(labels=('Heart No','Heart Yes'))
```
